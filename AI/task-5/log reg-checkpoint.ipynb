```python
data['age'].fillna(data['age'].mean(),inplace=True)
data['fare'].fillna(data['fare'].mean(),inplace=True)
```

```python
data['age'].isna().sum()
data['fare'].isna().sum()
```

```python
data.drop('cabin',axis=1,inplace=True)
data.drop('home.dest',axis=1,inplace=True)
data.drop('body',axis=1,inplace=True)
data.drop('boat',axis=1,inplace=True)
```

```python
data.head()
```

```python
pd.get_dummies(data['sex'])
```

```python
gender=pd.get_dummies(data['sex'],drop_first=True)
```

```python
data['gender']=gender
data.drop(['name','sex','ticket','embarked'],axis=1,inplace=True)
data.head()
```

```python
data.isna().sum()
```

```python
x=data[['passenger_id','pclass','age','sibsp','parch','fare','gender']]
y=data['survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
lr_m=LogisticRegression()
```

```python
lr_m.fit(x_train,y_train)
```

```python
predict=lr_m.predict(x_test)
```
